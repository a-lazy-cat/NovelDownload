```python
import requests
from bs4 import BeautifulSoup
from concurrent.futures import ThreadPoolExecutor

def get_content(url):
    response = requests.get(url)
    response.encoding = 'GB2312'
    soup = BeautifulSoup(response.text, 'lxml')
    texts = soup.find('div', id='content')
    newText = texts.text.replace('\r', '').replace('\n', '').replace(' www.23wx.vip ', '').strip().split('\xa0' * 4)
    return newText

def download_chapter(chapter):
    chapter_number, chapter_url = chapter
    content = get_content(chapter_url)
    return chapter_number, content

if __name__ == '__main__':
    target = 'https://www.23wx.vip/book_5690/'
    response = requests.get(target)
    response.encoding = 'GB2312'
    soup = BeautifulSoup(response.text, 'lxml')
    book_name = '.txt'
    chapters = soup.find('div', id='list').find_all('a')

    skip = 12  # Skip first 12 chapters
    chapters_info = [(i + 1, target + chapter.get('href')) for i, chapter in enumerate(chapters[skip:])]

    chapters_content = []
    with ThreadPoolExecutor(max_workers=10) as executor:
        futures = [executor.submit(download_chapter, chapter) for chapter in chapters_info]
        for future in futures:
            chapter_number, content = future.result()
            chapters_content.append((chapter_number, content))

    # Sort chapters by number
    chapters_content.sort(key=lambda x: x[0])

    # Write sorted chapters to file
    with open(book_name, 'a', encoding='utf-8') as f:
        for chapter_number, content in chapters_content:
            f.write(f"Chapter {chapter_number}")
            f.write('\n')
            f.write('\n'.join(content))
            f.write('\n')
```
